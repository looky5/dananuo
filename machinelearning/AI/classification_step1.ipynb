```python
import numpy as np
import pandas as pd
import re

product_data = pd.read_csv('./product.csv', error_bad_lines=False, dtype=str).fillna('없음')
```

### 몽클레어, 겐조, 에르노, 닐바렛 은(는) 모델명 이상으로 처음부터 제거

```python
idx_specific_brand = product_data[(product_data['제조사'] == '몽클레어') | (product_data['제조사'] == '겐조') | (product_data['제조사'] == '에르노') | (product_data['제조사'] == '닐바렛')].index
product_data = product_data.drop(idx_specific_brand)
```

```python
product_data = np.array(product_data[['상품 코드', '상품명', '중분류 카테고리명', '소분류 카테고리명', '세분류 카테고리명']])
```

```python
len(product_data)
```

```python
productmodelcode = []
p = re.compile(r'[a-zA-Z0-9-]{5,}')
p2 = re.compile(r'^[a-zA-Z-]{5,}$')
for single_product in product_data:
    matched = p.findall(single_product[1])
    if len(matched) != 0:
        justeng = p2.match(matched[-1])
        if justeng == None:
            productmodelcode.append([single_product[0], matched[-1], single_product[2], single_product[3], single_product[4]])
```

```python
productmodeltable = pd.DataFrame(productmodelcode, columns=['상품 코드', '모델명', '중분류 카테고리명', '소분류 카테고리명', '세분류 카테고리명'])
```

```python
productmodeltable
```

```python
productmodeltable_drop_dupl = productmodeltable.drop_duplicates(['모델명'], keep=False)
productmodeltable_drop_dupl
```

```python
df_product = productmodeltable_drop_dupl[['중분류 카테고리명', '소분류 카테고리명', '세분류 카테고리명']]
df_product = df_product.drop_duplicates(['중분류 카테고리명', '소분류 카테고리명', '세분류 카테고리명'])
```

```python
import time
```

```python
start = time.time()
linktable = []
for pdcategory in product_arr:
    omTable = df_openmarket_sample[(df_openmarket_sample['중분류 카테고리명'] == pdcategory[0]) & (df_openmarket_sample['소분류 카테고리명'] == pdcategory[1]) & (df_openmarket_sample['세분류 카테고리명'] == pdcategory[2])]
    omTabletoList = np.array(omTable['상품명'])
    pdTable = productmodeltable_drop_dupl[(productmodeltable_drop_dupl['중분류 카테고리명'] == pdcategory[0]) & (productmodeltable_drop_dupl['소분류 카테고리명'] == pdcategory[1]) & (productmodeltable_drop_dupl['세분류 카테고리명'] == pdcategory[2])]
    omlistseries = pd.Series(omTabletoList)
    for i in range(len(pdTable)):
        containsRes = omlistseries.str.contains(pdTable.iloc[i, 1], case=False)
        for j in range(len(containsRes)):
            if containsRes[j] == True:
                linktable.append([pdTable.iloc[i, 0], omTable.iloc[j, 0], omTable.iloc[j, 1]])
                df_openmarket_sample.drop(df_openmarket_sample.loc[(df_openmarket_sample['제휴사 코드'] == omTable.iloc[j, 0]) & (df_openmarket_sample['상품 코드'] == omTable.iloc[j, 1])].index, inplace=True)
print(time.time() - start)
```
