```python
start = time.time()

p = re.compile('[a-zA-Z]*[0-9]+[a-zA-Z0-9]*[-]?[_]?[s]?[a-zA-Z]*[0-9]+[a-zA-Z0-9]*[-]?[_]?[s]?[a-zA-Z]*[0-9]+[a-zA-Z0-9]*')
for i, product in enumerate (product_array):
    target = product[1].split(" ")
    new_str = []
    for kk in target:
        matchname = p.match(kk)
        if matchname == None:
            new_str.append(kk)
            new_str.append(" ")
    new_str = ''.join(new_str)
    new_str = new_str.strip()
#     print(product_array[i][1])
#     print(new_str)
    product_array[i][1] = new_str


openmarket_list = []
for i, openmarket in enumerate(openmarket_array):
    target = openmarket[2].split(" ")
#     print(target)
    new_str = []
    for kk in target:
        matchname = p.match(kk)
        if matchname == None:
            new_str.append(kk)
            new_str.append(" ")
    new_str = ''.join(new_str)
    new_str = new_str.strip()
#     print(openmarket_array[i][2])
#     print(new_str)
    openmarket_array[i][2] = new_str
    
print ('Cell took %.2f seconds to run.' %(time.time() - start))
```

```python
def blankSearch(name, nameLen, stack, blanklist, templist):
    for i in range(nameLen):
        if name[i] == '(':
            stack.append(i)
            templist.append(i)
        elif name[i] == ')' and len(stack) != 0:
            blanklist.append(stack.pop())
            blanklist.append(i)

    stack = []
    templist = []
    for i in range(nameLen):
        if name[i] == '[':
            stack.append(i)
            templist.append(i)
        elif name[i] == ']' and len(stack) != 0:
            blanklist.append(stack.pop())
            blanklist.append(i)
```

```python
start = time.time()

for i, openmarket in enumerate(openmarket_array):
    name = openmarket[2] # 상품명
    nameLen = len(name) # 상품명 길이

    # 괄호 구하기 위한 리스트
    stack = []
    blanklist = []
    templist = []
    
    blankSearch(name, nameLen, stack, blanklist, templist)
    # 정규식 사용하여 제품코드 추출하기
    blankLen = len(blanklist)
    blankLen = (int)(blankLen/2)
    linelist = []
    printlist = []
    for k in range(blankLen):   # 괄호만큼 돈다.
        pre = blanklist[k*2]
        post = blanklist[k*2+1]
        tempname = name[pre:post+1]
        printlist.append(tempname)
    for word in printlist:
        name = name.replace(word, " ")
    name = name.lstrip()
    openmarket_array[i][2] = name

print ('Cell took %.2f seconds to run.' %(time.time() - start))
```

### Correlation

```python
start = time.time()

result = []
for t_om in openmarket_array:
    openmarket_sentence = t_om[2].lower().split()
    p_o = len(openmarket_sentence)
    if p_o == 0:
        continue
    res_openmarketindex = t_om[1]
    res_openmarketname = t_om[2]
    res_productname = ""
    res_productidnex = -1
    res_omMiddle = t_om[6]
    res_omSmall = t_om[7]
    res_omDetail = t_om[8]
    cmax = 0
    pd_table = df_total_product[(df_total_product['중분류 카테고리명'] == res_omMiddle) & (df_total_product['소분류 카테고리명'] == res_omSmall) & (df_total_product['세분류 카테고리명'] == res_omDetail)]
    pd_array = pd_table.to_numpy()
    for t_p in pd_array:
        target_p = t_p[1].replace(" ","")
        p_p = 0

        for openmarket_word in openmarket_sentence:
            if openmarket_word in target_p:
                p_p += 1
        cor = p_p/p_o
        
        if(cmax<cor):
            res_productname = t_p[1]
            res_productindex = t_p[0]
            cmax = cor
            find = t_p[1]
            
    result.append([res_openmarketindex,res_openmarketname,res_productindex,res_productname, cmax])
print ('Cell took %.2f seconds to run.' %(time.time() - start))
```

```python
res_pcode = []
res_pname = []
res_check = []
for target in result:
    openmarket_code = target[0]
    link = df_total_link[df_total_link['제휴사 상품 코드'] == openmarket_code]
    product_code = link.iloc[0,2]
    pname = df_total_product[df_total_product['상품 코드'] == product_code]['상품명']
    res_pcode.append(product_code)
    res_pname.append(pname.iloc[0])
    if(target[2] == product_code):
        res_check.append(True)
    else:
        res_check.append(False)
```

```python
df_result = pd.DataFrame(result, columns=["외부 상품 코드","외부 상품명","비교 상품 코드","비교 상품명","조건부 확률"])
df_result.insert(4,'실 내부 코드',res_pcode)
df_result.insert(5,'실 내부 상품명',res_pname)
df_result.insert(6,'일치 여부',res_check)
```
